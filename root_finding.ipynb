### Newton's Method

We want to find a root of a function $f(x)$, i.e., solve:
$$
f(x) = 0
$$

Start with an initial guess $x_n$.  
Approximate $f(x)$ near $x_n$ using the first-order Taylor expansion:
$$
f(x) \approx f(x_n) + f'(x_n)(x - x_n)
$$

Set this approximation equal to zero:
$$
0 = f(x_n) + f'(x_n)(x - x_n)
$$

Solve for $x$:
$$
x = x_n - \frac{f(x_n)}{f'(x_n)}
$$

Define the next approximation:
$$
x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}
$$

This is Newton's iteration formula.



```python
# Newton's method

def newton_method(f, df, x0, tol=1e-10, max_iter=100):
    x = x0
    for i in range(max_iter):
        x_prev = x
        x = x - f(x) / df(x)
        print(f'iteration {i}: x = {x}')
        if abs(x - x_prev) < tol:
            break
    return x    
```

```python
# Find sqrt(2)
# f(x) = x^2 - 2
# f'(x) = 2x
# x_{n+1} = x_n - f(x_n) / f'(x_n)
# start with x_0 = 1.0

def f(x):
    return x**2 - 2

newton_method(f = f, df = lambda x: 2*x, x0 = 1.0)   
```

### Secant Method

Reall Newton's Method 

$$
x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}
$$

However, we don't have the analytical derivative , so instead we set

$$
f'(x_n) \approx \frac{f(x_n) - f(x_{n-1})}{x_n - x_{n-1}}
$$


```python
# Secant method

def secant_method(f, x0, x1, tol=1e-10, max_iter=100):
    for i in range(max_iter):
        deriv = (f(x0) - f(x1)) / (x0 - x1)
        x2 = x0 - f(x0) / deriv
        print(f'iteration {i}: x = {x2}')
        if abs(x2 - x0) < tol:
            break
        x1, x0 = x0, x2
    return x2
```

```python
# Find sqrt(2)
# f(x) = x^2 - 2
# x_{n+1} = x_n - f(x_n) / f'(x_n)
# f'(x) is approximated by (f(x_n) - f(x_{n-1})) / (x_n - x_{n-1})
# start with x_0 = 1.0

secant_method(f, 1.0, 1.1)
```
