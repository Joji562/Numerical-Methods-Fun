# The Midpoint Rule for Numerical Integration

## The Problem

We want to compute the definite integral:

$$I = \int_a^b f(x)\, dx$$

but $f$ may have no closed-form antiderivative, or we only know it at discrete points. The midpoint rule approximates $I$ by replacing $f$ with a piecewise-constant function whose height on each subinterval is sampled at the **midpoint**.

---

## Setup: Partition the Interval

Divide $[a, b]$ into $n$ equal subintervals of width:

$$h = \frac{b - a}{n}$$

The subinterval endpoints are:

$$x_i = a + i\,h, \quad i = 0, 1, \dots, n$$

so $x_0 = a$ and $x_n = b$. The **midpoint** of the $i$-th subinterval $[x_{i-1}, x_i]$ is:

$$m_i = x_{i-1} + \frac{h}{2} = a + \left(i - \frac{1}{2}\right)h, \quad i = 1, \dots, n$$

---

## The Approximation

On each subinterval $[x_{i-1}, x_i]$, approximate the integral by the area of a rectangle of width $h$ and height $f(m_i)$:

$$\int_{x_{i-1}}^{x_i} f(x)\,dx \approx h \cdot f(m_i)$$

Summing over all $n$ subintervals gives the **midpoint rule**:

$$\boxed{M_n = h \sum_{i=1}^{n} f(m_i) = \frac{b-a}{n} \sum_{i=1}^{n} f\!\left(a + \left(i - \tfrac{1}{2}\right)h\right)}$$

---

## Why Midpoint? A Taylor Series Proof

The key question is: why is the midpoint better than, say, the left or right endpoint?

Expand $f$ in a Taylor series around the midpoint $m_i$ of $[x_{i-1}, x_i]$:

$$f(x) = f(m_i) + f'(m_i)(x - m_i) + \frac{f''(m_i)}{2}(x - m_i)^2 + O\!\left((x-m_i)^3\right)$$

Integrate term by term over $[x_{i-1}, x_i]$:

$$\int_{x_{i-1}}^{x_i} f(x)\,dx = f(m_i)\cdot h + f'(m_i)\underbrace{\int_{x_{i-1}}^{x_i}(x - m_i)\,dx}_{=\; 0} + \frac{f''(m_i)}{2}\int_{x_{i-1}}^{x_i}(x-m_i)^2\,dx + \cdots$$

The **linear term vanishes exactly** because $m_i$ is the centre — the integral of $(x - m_i)$ over a symmetric interval is zero. So:

$$\int_{x_{i-1}}^{x_i} f(x)\,dx = h\,f(m_i) + \frac{f''(m_i)}{2} \cdot \frac{h^3}{12} + O(h^5)$$

since $\displaystyle\int_{m_i - h/2}^{m_i + h/2}(x - m_i)^2\,dx = \frac{h^3}{12}$.

The **local error** on one subinterval is therefore:

$$e_i = \frac{f''(\xi_i)}{24}\,h^3, \quad \xi_i \in [x_{i-1},\, x_i]$$

This is one order better than the left/right endpoint rules, which retain the linear term and have local error $O(h^2)$.

---

## Global Error

Summing the local error over all $n = (b-a)/h$ subintervals:

$$E_n = \sum_{i=1}^n e_i \approx n \cdot \frac{f''(\xi)}{24}\,h^3 = \frac{b-a}{24}\,f''(\xi)\,h^2$$

So the **global error** of the midpoint rule is:

$$\boxed{E_n = -\frac{(b-a)^3}{24\,n^2}\,f''(\xi), \quad \xi \in (a,b)}$$

The error scales as $O(h^2) = O(n^{-2})$: doubling $n$ cuts the error by a factor of **4**.

---

## Comparison of Newton–Cotes Rules

| Rule | Sample points | Local error | Global error |
|---|---|---|---|
| Left/Right endpoint | $f(x_{i-1})$ or $f(x_i)$ | $O(h^2)$ | $O(h)$ |
| **Midpoint** | $f(m_i)$ | $O(h^3)$ | $O(h^2)$ |
| Trapezoid | $f(x_{i-1}),\, f(x_i)$ | $O(h^3)$ | $O(h^2)$ |
| Simpson's | $f(x_{i-1}),\, f(m_i),\, f(x_i)$ | $O(h^5)$ | $O(h^4)$ |

The midpoint rule achieves **the same $O(h^2)$ global accuracy as the trapezoid rule**, despite using only one function evaluation per subinterval instead of two — making it twice as efficient per unit cost.

---

## Conditions and Caveats

| Concern | Detail |
|---|---|
| Requires $f \in C^2[a,b]$ | Error formula assumes $f''$ exists and is bounded |
| Singularities | If $f$ blows up inside $[a,b]$, accuracy degrades severely |
| Oscillatory $f$ | Need $h$ small enough to resolve the oscillation period |
| Only one sample per interval | Cannot detect sub-interval variation — use small $h$ |

```python
def midpoint_rule_integration(f, a, b, n):
    h = (b - a) / n
    total = 0
    for i in range(n):
        x_mid = a + (i + 0.5) * h
        total += f(x_mid)
    return total * h
```

# The Trapezoidal Rule for Numerical Integration

## The Problem

We want to compute:

$$I = \int_a^b f(x)\, dx$$

The trapezoidal rule approximates $f$ on each subinterval by a **straight line** connecting the two endpoints — a trapezoid rather than a rectangle.

---

## Setup: Partition the Interval

Divide $[a, b]$ into $n$ equal subintervals of width:

$$h = \frac{b - a}{n}$$

with nodes:

$$x_i = a + i\,h, \quad i = 0, 1, \dots, n$$

---

## The Approximation

On each subinterval $[x_{i-1}, x_i]$, replace $f$ with the **linear interpolant** through its two endpoints:

$$f(x) \approx f(x_{i-1}) + \frac{f(x_i) - f(x_{i-1})}{h}(x - x_{i-1})$$

Integrating this line exactly over $[x_{i-1}, x_i]$ gives the area of a trapezoid:

$$\int_{x_{i-1}}^{x_i} f(x)\,dx \approx \frac{h}{2}\bigl[f(x_{i-1}) + f(x_i)\bigr]$$

Summing over all $n$ subintervals:

$$T_n = \frac{h}{2}\sum_{i=1}^{n}\bigl[f(x_{i-1}) + f(x_i)\bigr]$$

Notice that every interior node $x_1, x_2, \dots, x_{n-1}$ appears **twice** — once as the right end of one trapezoid and once as the left end of the next. Collecting terms gives the compact form:

$$\boxed{T_n = \frac{h}{2}\Bigl[f(x_0) + 2f(x_1) + 2f(x_2) + \cdots + 2f(x_{n-1}) + f(x_n)\Bigr]}$$

or equivalently:

$$T_n = h\left[\frac{f(a)+f(b)}{2} + \sum_{i=1}^{n-1} f(x_i)\right]$$

The endpoints get weight $\tfrac{1}{2}$; every interior point gets weight $1$.

---

## Error Analysis: Taylor Expansion

Expand $f$ in a Taylor series around the **left** endpoint $x_{i-1}$:

$$f(x) = f(x_{i-1}) + f'(x_{i-1})(x - x_{i-1}) + \frac{f''(x_{i-1})}{2}(x-x_{i-1})^2 + O(h^3)$$

Integrating the exact $f$ over $[x_{i-1}, x_i]$:

$$\int_{x_{i-1}}^{x_i} f(x)\,dx = h\,f(x_{i-1}) + \frac{h^2}{2}f'(x_{i-1}) + \frac{h^3}{6}f''(x_{i-1}) + O(h^4)$$

Now expand the trapezoid approximation using Taylor at $x_{i-1}$:

$$\frac{h}{2}\bigl[f(x_{i-1}) + f(x_i)\bigr] = \frac{h}{2}\left[f(x_{i-1}) + f(x_{i-1}) + h f'(x_{i-1}) + \frac{h^2}{2}f''(x_{i-1}) + O(h^3)\right]$$

$$= h\,f(x_{i-1}) + \frac{h^2}{2}f'(x_{i-1}) + \frac{h^3}{4}f''(x_{i-1}) + O(h^4)$$

Subtracting the trapezoid from the exact integral gives the **local truncation error**:

$$e_i = \frac{h^3}{6}f''(x_{i-1}) - \frac{h^3}{4}f''(x_{i-1}) + O(h^4) = -\frac{h^3}{12}f''(\xi_i), \quad \xi_i \in [x_{i-1}, x_i]$$

---

## Why Does the Linear Term Survive?

Compare with the midpoint rule. There, the Taylor expansion around the **centre** $m_i$ caused the linear $(x - m_i)$ term to integrate to exactly zero by symmetry. Here we expand around an **endpoint**, so the linear term integrates to $h^2/2\,f'$ — which cancels only because the trapezoid formula itself contains the same $h^2/2\,f'$ term. The cancellation is algebraic, not geometric, and both methods end up with the same $O(h^3)$ local error.

---

## Global Error

Summing local errors over all $n = (b-a)/h$ subintervals and applying the mean value theorem for sums:

$$E_n = \sum_{i=1}^{n} e_i = -\frac{h^3}{12}\sum_{i=1}^{n}f''(\xi_i) \approx -\frac{h^3}{12} \cdot n \cdot f''(\xi)$$

Since $n = (b-a)/h$:

$$\boxed{E_n = -\frac{(b-a)}{12}f''(\xi)\,h^2 = -\frac{(b-a)^3}{12\,n^2}f''(\xi), \quad \xi \in (a,b)}$$

The global error is $O(h^2) = O(n^{-2})$: doubling $n$ reduces the error by a factor of **4**.

---

## The Euler–Maclaurin Perspective

The trapezoidal rule has a remarkable exact error expansion (Euler–Maclaurin formula):

$$\int_a^b f(x)\,dx = T_n - \frac{h^2}{12}\bigl[f'(b) - f'(a)\bigr] + \frac{h^4}{720}\bigl[f'''(b) - f'''(a)\bigr] - \cdots$$

Two consequences:

1. The error involves **only even powers of $h$** — so Richardson extrapolation (combining $T_n$ and $T_{2n}$) cleanly cancels the leading $h^2$ term and yields $O(h^4)$ accuracy. This is the basis of **Romberg integration**.

2. If $f$ is **periodic** and $[a,b]$ is exactly one period, all the boundary derivative terms vanish and the trapezoidal rule converges **exponentially fast** — far better than $O(h^2)$.

---

## Comparison with the Midpoint Rule

Both rules achieve $O(h^2)$ global accuracy, but their error constants have opposite signs:

$$E_n^{\text{trap}} = -\frac{(b-a)^3}{12\,n^2}f''(\xi), \qquad E_n^{\text{mid}} = +\frac{(b-a)^3}{24\,n^2}f''(\xi)$$

The midpoint rule's constant is **half** the size, making it roughly twice as accurate per function evaluation. Combining them with weights $\tfrac{1}{3}$ (trapezoid) and $\tfrac{2}{3}$ (midpoint) cancels the $h^2$ error entirely — this is exactly **Simpson's rule**:

$$S_n = \frac{1}{3}T_n + \frac{2}{3}M_n = \frac{h}{6}\bigl[f(x_{i-1}) + 4f(m_i) + f(x_i)\bigr]$$

achieving $O(h^4)$ global accuracy.

---

## Conditions and Caveats

| Concern | Detail |
|---|---|
| Requires $f \in C^2[a,b]$ | Error formula needs $f''$ bounded on $[a,b]$ |
| Concave $f$ | Trapezoids lie **below** the curve — rule underestimates |
| Convex $f$ | Trapezoids lie **above** the curve — rule overestimates |
| Singularities at endpoints | Accuracy degrades; consider substitution or Gauss quadrature |
| Periodic $f$ over full period | Exponential convergence — trapezoidal rule is near-optimal |

```python
def trapezoidal_rule_integration(f, a, b, n):
    h = (b - a) / n
    total = 0.5 * (f(a) + f(b))
    for i in range(1, n):
        x_i = a + i * h
        total += f(x_i)
    return total * h
```

# Simpson's Rule for Numerical Integration

## The Problem

We want to compute:

$$I = \int_a^b f(x)\, dx$$

Simpson's rule improves on the trapezoid rule by fitting a **quadratic (parabola)** through three equally-spaced points on each panel, rather than a straight line through two.

---

## Setup: Panels of Two Subintervals

We require $n$ to be **even**. Divide $[a, b]$ into $n$ subintervals of width:

$$h = \frac{b-a}{n}$$

with nodes $x_i = a + ih$. Group them into $n/2$ **panels**, each spanning two subintervals $[x_{2i-2},\, x_{2i}]$ with three nodes:

$$x_{2i-2} \quad (\text{left}), \qquad x_{2i-1} \quad (\text{midpoint}), \qquad x_{2i} \quad (\text{right})$$

---

## Deriving the Rule: Lagrange Interpolation

On a single panel $[-h, 0, h]$ (shifted to centre for clarity), fit the unique quadratic $p(x)$ through:

$$p(-h) = f_0, \quad p(0) = f_1, \quad p(h) = f_2$$

The Lagrange interpolating polynomial is:

$$p(x) = f_0 \frac{x(x-h)}{(-h)(-2h)} + f_1 \frac{(x+h)(x-h)}{h(-h)} + f_2 \frac{(x+h)x}{2h \cdot h}$$

Integrate $p(x)$ exactly over $[-h, h]$:

$$\int_{-h}^{h} p(x)\,dx = f_0 \cdot \frac{h}{3} + f_1 \cdot \frac{4h}{3} + f_2 \cdot \frac{h}{3}$$

$$= \frac{h}{3}\bigl[f_0 + 4f_1 + f_2\bigr]$$

This is the **Simpson's rule for one panel**: weight the endpoints by $1$, the midpoint by $4$, and multiply by $h/3$.

---

## The Full Composite Rule

Applying the one-panel formula to each of the $n/2$ panels and summing:

$$S_n = \frac{h}{3}\sum_{i=1}^{n/2}\bigl[f(x_{2i-2}) + 4f(x_{2i-1}) + f(x_{2i})\bigr]$$

Interior even-indexed nodes $x_2, x_4, \dots, x_{n-2}$ appear as the **right** end of one panel and the **left** end of the next — they are counted twice, gaining weight $1 + 1 = 2$. Interior odd-indexed nodes appear only as midpoints, keeping weight $4$. Collecting:

$$\boxed{S_n = \frac{h}{3}\Bigl[f(x_0) + 4f(x_1) + 2f(x_2) + 4f(x_3) + 2f(x_4) + \cdots + 4f(x_{n-1}) + f(x_n)\Bigr]}$$

The weight pattern is $1,\, 4,\, 2,\, 4,\, 2,\, \dots,\, 4,\, 1$.

---

## Simpson's Rule as a Weighted Average

Simpson's rule can be derived directly by combining the midpoint and trapezoid rules so their $O(h^2)$ errors cancel. Recall their error constants:

$$E^{\text{trap}} = -\frac{(b-a)}{12}f''(\xi)\,h^2, \qquad E^{\text{mid}} = +\frac{(b-a)}{24}f''(\xi)\,h^2$$

These have opposite signs and a $2:1$ ratio. Taking the weighted combination with weights $\tfrac{1}{3}$ and $\tfrac{2}{3}$:

$$S_n = \frac{1}{3}T_n + \frac{2}{3}M_n$$

The $h^2$ terms cancel exactly:

$$\frac{1}{3}\left(-\frac{h^2}{12}\right) + \frac{2}{3}\left(+\frac{h^2}{24}\right) = -\frac{h^2}{36} + \frac{h^2}{36} = 0$$

This is Richardson extrapolation in disguise, and it explains why Simpson's rule is so powerful — it costs only one extra function evaluation per panel (the midpoint) but gains **two full orders** of accuracy over the trapezoid rule.

---

## Error Analysis

For one panel $[x_{2i-2},\, x_{2i}]$, expand $f$ in a Taylor series around the panel midpoint $x_{2i-1}$ and integrate term by term. The quadratic interpolant $p(x)$ reproduces $f$ exactly up to degree 2. The error comes from the first term the parabola cannot capture:

$$\int_{x_{2i-2}}^{x_{2i}} f(x)\,dx - \frac{h}{3}\bigl[f_{2i-2} + 4f_{2i-1} + f_{2i}\bigr] = -\frac{h^5}{90}f^{(4)}(\xi_i), \quad \xi_i \in [x_{2i-2}, x_{2i}]$$

The local error is $O(h^5)$ — one order better than might be expected from fitting a degree-2 polynomial, because the cubic term also integrates to zero by the same symmetry argument as in the midpoint rule.

Summing over all $n/2$ panels with $n/2 = (b-a)/(2h)$:

$$\boxed{E_n = -\frac{(b-a)}{180}f^{(4)}(\xi)\,h^4 = -\frac{(b-a)^5}{180\,n^4}f^{(4)}(\xi), \quad \xi \in (a, b)}$$

The global error is $O(h^4) = O(n^{-4})$: doubling $n$ reduces the error by a factor of **16**.

---

## Why the Cubic Term Vanishes (the Bonus Order)

Fitting a quadratic gives exact results for polynomials up to degree 2 — one might expect $O(h^4)$ local error (from the cubic term). But expanding the local error around the panel midpoint $m$:

$$\int_{m-h}^{m+h}(x - m)^3\,dx = 0$$

The cubic term integrates to zero by **symmetry** — the interval is symmetric about $m$ and $(x-m)^3$ is an odd function. So the parabola also integrates the cubic term exactly, and the first nonzero error comes from the **quartic** term, giving local error $O(h^5)$ and global error $O(h^4)$.

This is the same symmetry trick that gave the midpoint rule a free order, now appearing again one level higher.

---

## Full Comparison of Newton–Cotes Rules

| Rule | Polynomial degree | Points per panel | Local error | Global error |
|---|---|---|---|---|
| Left/Right endpoint | 0 (constant) | 1 | $O(h^2)$ | $O(h)$ |
| Midpoint | 0 (constant, centred) | 1 | $O(h^3)$ | $O(h^2)$ |
| Trapezoid | 1 (linear) | 2 | $O(h^3)$ | $O(h^2)$ |
| **Simpson's** | **2 (quadratic)** | **3** | $O(h^5)$ | $O(h^4)$ |
| Simpson's 3/8 | 3 (cubic) | 4 | $O(h^5)$ | $O(h^4)$ |
| Boole's rule | 4 (quartic) | 5 | $O(h^7)$ | $O(h^6)$ |

Simpson's rule sits at the sweet spot: it uses only **3 points per panel** (one extra over the trapezoid) and gains two full orders of accuracy thanks to the odd-symmetry cancellation of the cubic term.

---

## Conditions and Caveats

| Concern | Detail |
|---|---|
| $n$ must be even | Panels consist of pairs of subintervals |
| Requires $f \in C^4[a,b]$ | Error formula needs $f^{(4)}$ bounded |
| Exact for cubics | Simpson integrates all polynomials up to degree 3 exactly |
| Singularities | Accuracy degrades near discontinuities or infinite derivatives |
| Oscillatory $f$ | Choose $h$ small enough so each panel captures the variation |

```python
def simpson_rule_integration(f, a, b, n):
    if n % 2 == 1:
        raise ValueError("n must be even for Simpson's rule")
    h = (b - a) / n
    total = f(a) + f(b)
    for i in range(1, n):
        x_i = a + i * h  
        if i % 2 == 0:
            total += 2 * f(x_i)
        else:
            total += 4 * f(x_i)
    return total * (h / 3)  
```

# Gaussian Quadrature

## The Core Idea

All the rules so far — midpoint, trapezoid, Simpson's — used **equally-spaced** nodes. Gaussian quadrature asks: what if we are free to choose **both** the node positions $x_i$ and the weights $w_i$? Can we do better?

The answer is dramatically yes. With $n$ free nodes and $n$ free weights ($2n$ free parameters total), we can construct a rule that integrates all polynomials of degree up to $2n - 1$ **exactly** — twice the degree achievable by any fixed-node rule with $n$ points.

---

## Setup: The Reference Interval

By a linear change of variables, any integral $\int_a^b f(x)\,dx$ maps to the standard interval $[-1, 1]$:

$$x = \frac{b-a}{2}t + \frac{a+b}{2}, \qquad dx = \frac{b-a}{2}\,dt$$

$$\int_a^b f(x)\,dx = \frac{b-a}{2}\int_{-1}^{1} f\!\left(\frac{b-a}{2}t + \frac{a+b}{2}\right)dt$$

So we focus on approximating $\displaystyle\int_{-1}^{1} g(t)\,dt$.

---

## The Approximation

An $n$-point Gaussian quadrature rule takes the form:

$$\int_{-1}^{1} g(t)\,dt \approx \sum_{i=1}^{n} w_i\, g(t_i)$$

where the **nodes** $t_1, \dots, t_n$ and **weights** $w_1, \dots, w_n$ are chosen optimally.

---

## Deriving the Nodes: Legendre Polynomials

We want the rule to be exact for all polynomials of degree $\leq 2n-1$. Write any such polynomial as:

$$p(t) = q(t)\,P_n(t) + r(t)$$

where $P_n(t)$ is the degree-$n$ **Legendre polynomial**, and $q(t)$, $r(t)$ have degree $\leq n-1$.

Legendre polynomials satisfy the orthogonality condition:

$$\int_{-1}^{1} P_n(t)\, q(t)\,dt = 0 \quad \text{for all polynomials } q \text{ of degree} < n$$

So:

$$\int_{-1}^{1} p(t)\,dt = \underbrace{\int_{-1}^{1} q(t)\,P_n(t)\,dt}_{=\;0} + \int_{-1}^{1} r(t)\,dt = \int_{-1}^{1} r(t)\,dt$$

Choose the **nodes $t_i$ to be the $n$ roots of $P_n(t)$**. Then $P_n(t_i) = 0$ for all $i$, so:

$$\sum_{i=1}^{n} w_i\, p(t_i) = \sum_{i=1}^{n} w_i \underbrace{q(t_i)\,P_n(t_i)}_{=\;0} + \sum_{i=1}^{n} w_i\, r(t_i) = \sum_{i=1}^{n} w_i\, r(t_i)$$

Since $r$ has degree $\leq n-1$, any set of weights that integrates degree-$(n-1)$ polynomials exactly will handle $r$ perfectly. The weights are chosen to do exactly this via the **Lagrange basis**:

$$w_i = \int_{-1}^{1} \ell_i(t)\,dt, \qquad \ell_i(t) = \prod_{j \neq i} \frac{t - t_j}{t_i - t_j}$$

This guarantees exact integration of all polynomials of degree $\leq 2n - 1$.

---

## The First Few Gauss–Legendre Rules

**$n = 1$** (exact for degree $\leq 1$):

$$\int_{-1}^{1} g(t)\,dt \approx 2\,g(0), \qquad t_1 = 0,\quad w_1 = 2$$

**$n = 2$** (exact for degree $\leq 3$):

$$\int_{-1}^{1} g(t)\,dt \approx g\!\left(-\tfrac{1}{\sqrt{3}}\right) + g\!\left(\tfrac{1}{\sqrt{3}}\right)$$

$$t_{1,2} = \pm\frac{1}{\sqrt{3}} \approx \pm 0.5774, \qquad w_1 = w_2 = 1$$

**$n = 3$** (exact for degree $\leq 5$):

$$\int_{-1}^{1} g(t)\,dt \approx \frac{5}{9}\,g\!\left(-\sqrt{\tfrac{3}{5}}\right) + \frac{8}{9}\,g(0) + \frac{5}{9}\,g\!\left(\sqrt{\tfrac{3}{5}}\right)$$

$$t_{1,3} = \pm\sqrt{\tfrac{3}{5}} \approx \pm 0.7746,\quad w_{1,3} = \tfrac{5}{9} \approx 0.556$$
$$t_2 = 0,\quad w_2 = \tfrac{8}{9} \approx 0.889$$

---

## Applying to a General Interval $[a, b]$

Using the substitution $t = \frac{2x - (a+b)}{b-a}$, the full rule becomes:

$$\boxed{\int_a^b f(x)\,dx \approx \frac{b-a}{2}\sum_{i=1}^{n} w_i\, f\!\left(\frac{b-a}{2}t_i + \frac{a+b}{2}\right)}$$

---

## Error Analysis

For an $n$-point Gauss–Legendre rule applied to a smooth function:

$$E_n = \frac{(b-a)^{2n+1}(n!)^4}{(2n+1)\bigl[(2n)!\bigr]^3} f^{(2n)}(\xi), \qquad \xi \in (a, b)$$

The key feature is that the error depends on the **$2n$-th derivative** of $f$ — meaning Gaussian quadrature achieves $O(h^{2n})$ accuracy with only $n$ function evaluations. Compare this to:

$$\text{Simpson (n=3 points)} \rightarrow O(h^4), \qquad \text{Gauss (n=3 points)} \rightarrow O(h^6)$$

For smooth functions, convergence is effectively **exponential** as $n$ grows — the number of correct digits doubles (roughly) with each additional node.

---

## Why Nodes Are Not Equally Spaced

Equally-spaced nodes suffer from **Runge's phenomenon**: high-degree polynomial interpolation over uniform nodes develops large oscillations near the endpoints, destroying accuracy. The Gauss nodes are the roots of Legendre polynomials, which cluster toward $\pm 1$:

$$t_i \approx \cos\!\left(\frac{2i-1}{2n}\pi\right)$$

This clustering is not accidental — it is the same distribution as Chebyshev nodes, which are known to minimise the maximum interpolation error. The endpoints themselves are never nodes, which avoids the worst oscillations.

---

## Comparison of Quadrature Rules

| Rule | $n$ points | Degree exact | Global error | Notes |
|---|---|---|---|---|
| Midpoint | $n$ | 1 | $O(h^2)$ | Fixed nodes |
| Trapezoid | $n$ | 1 | $O(h^2)$ | Fixed nodes |
| Simpson's | $n$ (even) | 3 | $O(h^4)$ | Fixed nodes |
| **Gauss ($k$ pts/panel)** | $n$ | $2k-1$ | $O(h^{2k})$ | **Optimal nodes** |

With $k=2$ Gauss points per panel Simpson's degree (3) is matched using fewer evaluations; with $k=3$ points per panel you surpass Boole's rule.

---

## Conditions and Caveats

| Concern | Detail |
|---|---|
| Smoothness required | Error formula assumes $f^{(2n)}$ exists and is bounded |
| Not adaptive | Nodes are fixed; singularities inside $(a,b)$ ruin accuracy |
| Lookup tables | Nodes and weights must be precomputed (or looked up) for each $n$ |
| Endpoint singularities | Gauss–Jacobi and Gauss–Laguerre variants handle weighted integrals $\int w(x)f(x)\,dx$ |
| Composite version | Split $[a,b]$ into panels and apply low-order Gauss rule per panel for robustness |

# Adaptive Quadrature

## Motivation

All previous rules — Simpson's, Gaussian, etc. — use a **fixed** set of nodes across $[a, b]$.
This is wasteful when $f$ varies wildly in one region and smoothly in another:
- A coarse grid over a smooth region is accurate enough.
- The same coarse grid over a spike or near-singularity can be catastrophically wrong.

**Adaptive quadrature** allocates nodes where they are needed, automatically, by monitoring a local error estimate and subdividing only the panels that fail a tolerance test.

---

## The Core Principle: Local Error Estimation

Apply two quadrature rules of different orders to the same interval $[a, b]$:

- A **coarse** estimate $Q_1$ (e.g. Simpson's with one panel)
- A **fine** estimate $Q_2$ (e.g. Simpson's with two panels)

Use their difference as a cheap error estimate:

$$\widehat{E} = |Q_2 - Q_1|$$

If $\widehat{E} < \varepsilon_{\text{local}}$, accept $Q_2$ as the integral over $[a,b]$.
If $\widehat{E} \geq \varepsilon_{\text{local}}$, **bisect**: split $[a,b]$ at the midpoint $m = \frac{a+b}{2}$ and recurse on each half.

---

## The Algorithm (Recursive Form)

Define $\texttt{adapt}(f, a, b, \varepsilon, Q_{\text{whole}})$:

**1.** Compute $m = \dfrac{a + b}{2}$.

**2.** Compute left and right sub-estimates:
$$Q_L = S(f, a, m), \qquad Q_R = S(f, m, b)$$
where $S(f, c, d)$ denotes Simpson's rule on $[c, d]$.

**3.** Estimate error:
$$\widehat{E} = |Q_L + Q_R - Q_{\text{whole}}|$$

**4.** If $\widehat{E} < 15\,\varepsilon$:

$$\text{return } Q_L + Q_R + \frac{Q_L + Q_R - Q_{\text{whole}}}{15}$$

Otherwise, recurse:

$$\text{return } \texttt{adapt}\!\left(f, a, m, \tfrac{\varepsilon}{2}, Q_L\right) + \texttt{adapt}\!\left(f, m, b, \tfrac{\varepsilon}{2}, Q_R\right)$$

The tolerance is **halved** at each subdivision so that errors across all sub-panels sum to $\leq \varepsilon$ globally.

---

## Why the Factor of 15?

When bisecting, Simpson's rule error scales as $h^4$. Halving the interval reduces the error on each half by $2^4 = 16$, so the combined error over both halves is $\frac{1}{16}$ of the original. The improvement is:

$$Q_{\text{whole}} - (Q_L + Q_R) \approx E_{\text{whole}} - 2 \cdot \frac{E_{\text{whole}}}{16} = E_{\text{whole}}\left(1 - \frac{1}{8}\right) = \frac{15}{16}\,E_{\text{whole}}$$

Rearranging gives the **Richardson-corrected** estimate:

$$\int_a^b f\,dx \approx Q_L + Q_R + \underbrace{\frac{Q_L + Q_R - Q_{\text{whole}}}{15}}_{\text{error correction}}$$

and a reliable stopping criterion:

$$|Q_L + Q_R - Q_{\text{whole}}| < 15\,\varepsilon$$

---

## Error Propagation and the $\varepsilon/2$ Splitting

At each bisection the tolerance is split equally between the two children. For a binary recursion tree of depth $d$, there are $2^d$ leaves each with tolerance $\varepsilon / 2^d$. The total error bound is:

$$E_{\text{total}} \leq \sum_{\text{leaves}} \frac{\varepsilon}{2^d} \cdot 2^d \cdot \frac{1}{2^d} \longrightarrow \varepsilon$$

More precisely, if the recursion terminates with $k$ accepted leaf panels $[a_j, b_j]$ each satisfying $\widehat{E}_j < 15\,\varepsilon_j$, then:

$$\left|\int_a^b f\,dx - \sum_{j=1}^{k} (Q_L^j + Q_R^j)\right| \leq \sum_{j=1}^k \varepsilon_j = \varepsilon$$

provided the tolerance is split so that $\sum_j \varepsilon_j = \varepsilon$.

---

## Comparison: Uniform vs Adaptive

For a function with a sharp feature of width $\delta$ inside a smooth background on $[a, b]$:

**Uniform Simpson's** to achieve global error $\varepsilon$ requires:

$$n \sim \left(\frac{(b-a)^5 \|f^{(4)}\|_\infty}{180\,\varepsilon}\right)^{1/4}$$

which is dominated by $\|f^{(4)}\|_\infty$ — potentially enormous near a spike.

**Adaptive quadrature** concentrates panels of width $\sim \delta$ only near the feature. The number of evaluations scales as:

$$n_{\text{adapt}} \sim \underbrace{C_{\text{smooth}}}_{\text{few panels}} + \underbrace{\frac{\delta^{1/4}}{\varepsilon^{1/4}} \cdot C_{\text{spike}}}_{\text{many panels near spike}}$$

For $\delta \ll (b-a)$ this is orders of magnitude cheaper.

## Practical Considerations

### Absolute vs relative tolerance

In practice, two tolerances are specified:

$$\widehat{E} < \varepsilon_{\text{abs}} + \varepsilon_{\text{rel}}\,|Q|$$

This prevents the stopping criterion from being too strict when the integral itself is very small.

### Singularities

If $f$ has an **integrable singularity** (e.g. $f(x) \sim |x - c|^{-1/2}$), adaptive quadrature still converges — the panels near $c$ become exponentially small. For **non-integrable singularities**, the recursion never terminates and a max-depth guard is essential.

### Stack depth

Naive recursion can overflow the call stack for very spiky functions. Production implementations (e.g. `scipy.integrate.quad`, which wraps QUADPACK) use an explicit priority queue, always refining the panel with the **largest estimated error** first.

### Reuse of function evaluations

When a panel $[a, b]$ is bisected, the five points already evaluated for Simpson's rule on $[a, b]$ include the endpoints and midpoint. The two children each need two new interior points, so each bisection costs only **4 new evaluations**, not 10.

## Connection to Previous Rules

Adaptive quadrature is not a new integration formula — it is a **strategy** for applying existing formulas intelligently:

$$\text{Adaptive} = \text{Simpson (or Gauss)} + \text{recursive bisection} + \text{Richardson error estimate}$$

The building blocks from earlier notes slot directly in:

| Component | Source |
|---|---|
| One-panel estimate $S(f, a, b)$ | Simpson's rule |
| Error order $O(h^4)$ | Simpson's error analysis |
| Factor of 15 denominator | Richardson extrapolation ($2^4 - 1 = 15$) |
| Optimal node placement per panel | Gauss–Legendre (in production codes) |

## Conditions and Caveats

| Concern | Detail |
|---|---|
| Oscillatory $f$ | May accept a panel that looks smooth but contains hidden oscillations — use a minimum panel count |
| Discontinuities | Bisection stalls near a jump; bracket and split at the discontinuity manually |
| High-dimensional integrals | Adaptive 1D quadrature does not extend cheaply to $\mathbb{R}^d$ — use Monte Carlo instead |
| Max recursion depth | Always set a hard limit and raise a warning if hit |
| Cost model | Each bisection costs 4 new evaluations; budget evaluations, not panels |
