# The Midpoint Rule for Numerical Integration

## The Problem

We want to compute the definite integral:

$$I = \int_a^b f(x)\, dx$$

but $f$ may have no closed-form antiderivative, or we only know it at discrete points. The midpoint rule approximates $I$ by replacing $f$ with a piecewise-constant function whose height on each subinterval is sampled at the **midpoint**.

---

## Setup: Partition the Interval

Divide $[a, b]$ into $n$ equal subintervals of width:

$$h = \frac{b - a}{n}$$

The subinterval endpoints are:

$$x_i = a + i\,h, \quad i = 0, 1, \dots, n$$

so $x_0 = a$ and $x_n = b$. The **midpoint** of the $i$-th subinterval $[x_{i-1}, x_i]$ is:

$$m_i = x_{i-1} + \frac{h}{2} = a + \left(i - \frac{1}{2}\right)h, \quad i = 1, \dots, n$$

---

## The Approximation

On each subinterval $[x_{i-1}, x_i]$, approximate the integral by the area of a rectangle of width $h$ and height $f(m_i)$:

$$\int_{x_{i-1}}^{x_i} f(x)\,dx \approx h \cdot f(m_i)$$

Summing over all $n$ subintervals gives the **midpoint rule**:

$$\boxed{M_n = h \sum_{i=1}^{n} f(m_i) = \frac{b-a}{n} \sum_{i=1}^{n} f\!\left(a + \left(i - \tfrac{1}{2}\right)h\right)}$$

---

## Why Midpoint? A Taylor Series Proof

The key question is: why is the midpoint better than, say, the left or right endpoint?

Expand $f$ in a Taylor series around the midpoint $m_i$ of $[x_{i-1}, x_i]$:

$$f(x) = f(m_i) + f'(m_i)(x - m_i) + \frac{f''(m_i)}{2}(x - m_i)^2 + O\!\left((x-m_i)^3\right)$$

Integrate term by term over $[x_{i-1}, x_i]$:

$$\int_{x_{i-1}}^{x_i} f(x)\,dx = f(m_i)\cdot h + f'(m_i)\underbrace{\int_{x_{i-1}}^{x_i}(x - m_i)\,dx}_{=\; 0} + \frac{f''(m_i)}{2}\int_{x_{i-1}}^{x_i}(x-m_i)^2\,dx + \cdots$$

The **linear term vanishes exactly** because $m_i$ is the centre — the integral of $(x - m_i)$ over a symmetric interval is zero. So:

$$\int_{x_{i-1}}^{x_i} f(x)\,dx = h\,f(m_i) + \frac{f''(m_i)}{2} \cdot \frac{h^3}{12} + O(h^5)$$

since $\displaystyle\int_{m_i - h/2}^{m_i + h/2}(x - m_i)^2\,dx = \frac{h^3}{12}$.

The **local error** on one subinterval is therefore:

$$e_i = \frac{f''(\xi_i)}{24}\,h^3, \quad \xi_i \in [x_{i-1},\, x_i]$$

This is one order better than the left/right endpoint rules, which retain the linear term and have local error $O(h^2)$.

---

## Global Error

Summing the local error over all $n = (b-a)/h$ subintervals:

$$E_n = \sum_{i=1}^n e_i \approx n \cdot \frac{f''(\xi)}{24}\,h^3 = \frac{b-a}{24}\,f''(\xi)\,h^2$$

So the **global error** of the midpoint rule is:

$$\boxed{E_n = -\frac{(b-a)^3}{24\,n^2}\,f''(\xi), \quad \xi \in (a,b)}$$

The error scales as $O(h^2) = O(n^{-2})$: doubling $n$ cuts the error by a factor of **4**.

---

## Comparison of Newton–Cotes Rules

| Rule | Sample points | Local error | Global error |
|---|---|---|---|
| Left/Right endpoint | $f(x_{i-1})$ or $f(x_i)$ | $O(h^2)$ | $O(h)$ |
| **Midpoint** | $f(m_i)$ | $O(h^3)$ | $O(h^2)$ |
| Trapezoid | $f(x_{i-1}),\, f(x_i)$ | $O(h^3)$ | $O(h^2)$ |
| Simpson's | $f(x_{i-1}),\, f(m_i),\, f(x_i)$ | $O(h^5)$ | $O(h^4)$ |

The midpoint rule achieves **the same $O(h^2)$ global accuracy as the trapezoid rule**, despite using only one function evaluation per subinterval instead of two — making it twice as efficient per unit cost.

---

## Conditions and Caveats

| Concern | Detail |
|---|---|
| Requires $f \in C^2[a,b]$ | Error formula assumes $f''$ exists and is bounded |
| Singularities | If $f$ blows up inside $[a,b]$, accuracy degrades severely |
| Oscillatory $f$ | Need $h$ small enough to resolve the oscillation period |
| Only one sample per interval | Cannot detect sub-interval variation — use small $h$ |

```python
def midpoint_rule_integration(f, a, b, n):
    h = (b - a) / n
    total = 0
    for i in range(n):
        x_mid = a + (i + 0.5) * h
        total += f(x_mid)
    return total * h
```

# The Trapezoidal Rule for Numerical Integration

## The Problem

We want to compute:

$$I = \int_a^b f(x)\, dx$$

The trapezoidal rule approximates $f$ on each subinterval by a **straight line** connecting the two endpoints — a trapezoid rather than a rectangle.

---

## Setup: Partition the Interval

Divide $[a, b]$ into $n$ equal subintervals of width:

$$h = \frac{b - a}{n}$$

with nodes:

$$x_i = a + i\,h, \quad i = 0, 1, \dots, n$$

---

## The Approximation

On each subinterval $[x_{i-1}, x_i]$, replace $f$ with the **linear interpolant** through its two endpoints:

$$f(x) \approx f(x_{i-1}) + \frac{f(x_i) - f(x_{i-1})}{h}(x - x_{i-1})$$

Integrating this line exactly over $[x_{i-1}, x_i]$ gives the area of a trapezoid:

$$\int_{x_{i-1}}^{x_i} f(x)\,dx \approx \frac{h}{2}\bigl[f(x_{i-1}) + f(x_i)\bigr]$$

Summing over all $n$ subintervals:

$$T_n = \frac{h}{2}\sum_{i=1}^{n}\bigl[f(x_{i-1}) + f(x_i)\bigr]$$

Notice that every interior node $x_1, x_2, \dots, x_{n-1}$ appears **twice** — once as the right end of one trapezoid and once as the left end of the next. Collecting terms gives the compact form:

$$\boxed{T_n = \frac{h}{2}\Bigl[f(x_0) + 2f(x_1) + 2f(x_2) + \cdots + 2f(x_{n-1}) + f(x_n)\Bigr]}$$

or equivalently:

$$T_n = h\left[\frac{f(a)+f(b)}{2} + \sum_{i=1}^{n-1} f(x_i)\right]$$

The endpoints get weight $\tfrac{1}{2}$; every interior point gets weight $1$.

---

## Error Analysis: Taylor Expansion

Expand $f$ in a Taylor series around the **left** endpoint $x_{i-1}$:

$$f(x) = f(x_{i-1}) + f'(x_{i-1})(x - x_{i-1}) + \frac{f''(x_{i-1})}{2}(x-x_{i-1})^2 + O(h^3)$$

Integrating the exact $f$ over $[x_{i-1}, x_i]$:

$$\int_{x_{i-1}}^{x_i} f(x)\,dx = h\,f(x_{i-1}) + \frac{h^2}{2}f'(x_{i-1}) + \frac{h^3}{6}f''(x_{i-1}) + O(h^4)$$

Now expand the trapezoid approximation using Taylor at $x_{i-1}$:

$$\frac{h}{2}\bigl[f(x_{i-1}) + f(x_i)\bigr] = \frac{h}{2}\left[f(x_{i-1}) + f(x_{i-1}) + h f'(x_{i-1}) + \frac{h^2}{2}f''(x_{i-1}) + O(h^3)\right]$$

$$= h\,f(x_{i-1}) + \frac{h^2}{2}f'(x_{i-1}) + \frac{h^3}{4}f''(x_{i-1}) + O(h^4)$$

Subtracting the trapezoid from the exact integral gives the **local truncation error**:

$$e_i = \frac{h^3}{6}f''(x_{i-1}) - \frac{h^3}{4}f''(x_{i-1}) + O(h^4) = -\frac{h^3}{12}f''(\xi_i), \quad \xi_i \in [x_{i-1}, x_i]$$

---

## Why Does the Linear Term Survive?

Compare with the midpoint rule. There, the Taylor expansion around the **centre** $m_i$ caused the linear $(x - m_i)$ term to integrate to exactly zero by symmetry. Here we expand around an **endpoint**, so the linear term integrates to $h^2/2\,f'$ — which cancels only because the trapezoid formula itself contains the same $h^2/2\,f'$ term. The cancellation is algebraic, not geometric, and both methods end up with the same $O(h^3)$ local error.

---

## Global Error

Summing local errors over all $n = (b-a)/h$ subintervals and applying the mean value theorem for sums:

$$E_n = \sum_{i=1}^{n} e_i = -\frac{h^3}{12}\sum_{i=1}^{n}f''(\xi_i) \approx -\frac{h^3}{12} \cdot n \cdot f''(\xi)$$

Since $n = (b-a)/h$:

$$\boxed{E_n = -\frac{(b-a)}{12}f''(\xi)\,h^2 = -\frac{(b-a)^3}{12\,n^2}f''(\xi), \quad \xi \in (a,b)}$$

The global error is $O(h^2) = O(n^{-2})$: doubling $n$ reduces the error by a factor of **4**.

---

## The Euler–Maclaurin Perspective

The trapezoidal rule has a remarkable exact error expansion (Euler–Maclaurin formula):

$$\int_a^b f(x)\,dx = T_n - \frac{h^2}{12}\bigl[f'(b) - f'(a)\bigr] + \frac{h^4}{720}\bigl[f'''(b) - f'''(a)\bigr] - \cdots$$

Two consequences:

1. The error involves **only even powers of $h$** — so Richardson extrapolation (combining $T_n$ and $T_{2n}$) cleanly cancels the leading $h^2$ term and yields $O(h^4)$ accuracy. This is the basis of **Romberg integration**.

2. If $f$ is **periodic** and $[a,b]$ is exactly one period, all the boundary derivative terms vanish and the trapezoidal rule converges **exponentially fast** — far better than $O(h^2)$.

---

## Comparison with the Midpoint Rule

Both rules achieve $O(h^2)$ global accuracy, but their error constants have opposite signs:

$$E_n^{\text{trap}} = -\frac{(b-a)^3}{12\,n^2}f''(\xi), \qquad E_n^{\text{mid}} = +\frac{(b-a)^3}{24\,n^2}f''(\xi)$$

The midpoint rule's constant is **half** the size, making it roughly twice as accurate per function evaluation. Combining them with weights $\tfrac{1}{3}$ (trapezoid) and $\tfrac{2}{3}$ (midpoint) cancels the $h^2$ error entirely — this is exactly **Simpson's rule**:

$$S_n = \frac{1}{3}T_n + \frac{2}{3}M_n = \frac{h}{6}\bigl[f(x_{i-1}) + 4f(m_i) + f(x_i)\bigr]$$

achieving $O(h^4)$ global accuracy.

---

## Conditions and Caveats

| Concern | Detail |
|---|---|
| Requires $f \in C^2[a,b]$ | Error formula needs $f''$ bounded on $[a,b]$ |
| Concave $f$ | Trapezoids lie **below** the curve — rule underestimates |
| Convex $f$ | Trapezoids lie **above** the curve — rule overestimates |
| Singularities at endpoints | Accuracy degrades; consider substitution or Gauss quadrature |
| Periodic $f$ over full period | Exponential convergence — trapezoidal rule is near-optimal |

```python
def trapezoidal_rule_integration(f, a, b, n):
    h = (b - a) / n
    total = 0.5 * (f(a) + f(b))
    for i in range(1, n):
        x_i = a + i * h
        total += f(x_i)
    return total * h
```

```python
def simpson_rule_integration(f, a, b, n):
    if n % 2 == 1:
        raise ValueError("n must be even for Simpson's rule")
    h = (b - a) / n
    total = f(a) + f(b)
    for i in range(1, n):
        x_i = a + i * h
        if i % 2 == 0:
            total += 2 * f(x_i)
        else:
            total += 4 * f(x_i)
    return total * (h / 3)  
```
